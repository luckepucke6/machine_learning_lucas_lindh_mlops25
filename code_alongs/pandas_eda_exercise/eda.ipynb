```python
import pandas as pd
```

```python
df = pd.read_csv("Sales.csv")
```

```python
df.info()
df
```

```python
df["Year"].value_counts()
```

```python
revenue_per_year = df.groupby("Year")["Revenue"].sum()
revenue_per_year
```

```python
revenue_per_year.plot(kind="bar", x="Year", y="Revenue", title="Revenue per year")
```

```python
country = df["Country"].value_counts()
```

```python
country.plot(kind="bar", y="Sales", x="Country", title="Sales by country")
```

```python
country = df.groupby("Product_Category")["Country"].value_counts().reset_index(name="Sales")
country
```

```python
country.plot(kind="bar", x="Product_Category", y="Sales")
```
