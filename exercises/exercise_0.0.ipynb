## d)

```python
df
```

```python
import_by_year_month = df.groupby(["Year", "Month"])["Import"].sum().reset_index()

import_by_year_month["Date"] = pd.to_datetime(
    import_by_year_month["Year"].astype(str)
    + "-"
    + import_by_year_month["Month"].astype(str)
)

import_by_year_month = import_by_year_month.sort_values("Date")

plt.plot(import_by_year_month["Date"], import_by_year_month["Import"])
plt.xlabel("Year/Month")
plt.ylabel("Import")
plt.title("Import over time")
plt.show()
```

```python
avg_co2_by_year_month = df.groupby(["Year", "Month"])["Avg_CO2"].sum().reset_index()

avg_co2_by_year_month["Date"] = pd.to_datetime(
    avg_co2_by_year_month["Year"].astype(str)
    + "-"
    + avg_co2_by_year_month["Month"].astype(str)
)

avg_co2_by_year_month = avg_co2_by_year_month.sort_values("Date")

plt.plot(avg_co2_by_year_month["Date"], avg_co2_by_year_month["Avg_CO2"])
plt.xlabel("Year/Month")
plt.ylabel("Avg_CO2")
plt.title("Avg_CO2 over time")
plt.show()
```

## f)

```python
df.groupby("Year")["Import_Electric"].sum()
```

```python
mask = df[(df["Year"] >= 2012) & (df["Import_Electric"].notna())]
```

```python
mask["Year"].unique()
mask = mask.sort_values(by=["Year", "Month"])
```

```python
mask["Date"] = pd.to_datetime(
    mask["Year"].astype(str) + "-" + mask["Month"].astype(str)
)

plt.figure(figsize=(10, 5))
plt.plot(mask["Date"], mask["Import_Electric"])
plt.xlabel("Date")
plt.ylabel("Import Electric")
```
