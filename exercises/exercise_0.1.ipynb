## d)

```python
salary_summary = df.groupby(["job_title", "experience_level"])["salary_in_usd"].agg(
    ["median", "mean"]
)
```

```python
salary_summary.reset_index()
```

```python
salary_summary = salary_summary.rename(
    columns={"median": "median_salary_usd", "mean": "mean_salary_usd"}
)
```

```python
salary_summary["mean_salary_sek"] = (salary_summary["mean_salary_usd"] * 10.5)
```

```python
salary_summary.sort_values(by="mean_salary_sek", ascending=False)
```

```python
salary_summary.head(20)
```

```python
df["salary_sek_per_month"] = (df["salary_in_usd"] * 10.5) / 12
df
```

```python
plt.figure(figsize=(8,4))
plt.hist(df["salary_sek_per_month"], bins=50)

plt.title("Distribution of monthly salaries in SEK")
plt.xlabel("Monthly salary (SEK)")
plt.ylabel("Number of job postings")

plt.tight_layout()
```
